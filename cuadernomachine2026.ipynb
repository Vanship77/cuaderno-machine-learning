*   **Referencias a objetos:**
    *   Las variables y los objetos se almacenan en diferentes zonas de memoria.

    *   Objetos sí que pueden referenciar a otros objetos. Ejemplo: listas.

*   **Sentencia de asignación:**
    <!-- FORMATO A SEGUIR: <nombre_variable> '=' <objeto> -->

```python
 # # Asignación de variables
a = 5
print(a)
```

```python
a = 1 # entero
b = 4.0 # coma flotante
c = "ITQ" # string
d = 10 + 1j # numero complejo
e = True #False # boolean
f = None # None
# visualizar valor de las variables y su tipo
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
print(d)
print(type(d))
print(e)
print(type(e))
print(f)
print(type(f))
```

<h2>Identificadores</h2>

* Podemos obtener un identificador único para los objetos referenciados por variables.
* Este identificador se obtiene a partir de la dirección de memoria.

```python
a = 3
print(id(a))
a = 'Pablo García'
print(id(a))
a = 4.5
print(id(a))
```

* Referencias compartidas: un mismo objeto puede ser referenciado por más de una variable.
* Variables que referencian al mismo objeto tienen mismo identificador.


```python
a = 4567
print(id(a))
```

```python
b = a
print(id(b))
```

```python
c = 4567
print(id(c))
```

```python
a = 25
b = 25
print(id(a))
print(id(b))
print(id(25))
```

* Union, interseccion y diferencia

```python
set1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set2 = set([0, 1, 2, 3, 4, 42])
# union
print(set1 | set2)
# intersección
print(set1 & set2)
# diferencia
print(set1 - set2)
print(set2 - set1)
```

```python
a = [1,2]
b = [2,3]
print(set(a) & set(b))
```

```python
#Ademas de los operadores, que operan únicamente con Sets, también se pueden usar metodos que pueden operar sobre cualquier objeto iterable.
conjunto = {0, 1, 2}
lista = [1, 3, 3]
print(conjunto.union(lista))
print(conjunto.intersection(lista))
print(conjunto.difference(lista))
```

```python
set1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set2 = set([0, 1, 2, 3, 4])
print(set2.issubset(set1))
print(set1.issuperset(set2))
print(set1.isdisjoint(set2))
```

* Modificación de conjuntos

```python
# A través de operador de asignación
set1 = {'a', 'b', 'c'}
set2 = {'a', 'd'}
# set1 |= set2 # set1 = set1 | set2
# set1 &= set2
set1 -= set2
print(set1)
```

```python
# A través de método 'update'.
set1 = {'a', 'b', 'c'}
set2 = {'a', 'd'}
# set1.update(set2)
# set1.intersection_update(set2)
set1.difference_update(set2)
print(set1)
```

```python
# A través de métodos 'add' y 'remove'.
set1 = {'a', 'b', 'c'}
set1.add('d')
set1.remove('a')
print(set1)
```

<h2>3.4 Tuplas</h2>

<h4>Al igual que las listas:</h4>
<ul>
  <li>Colección de elementos.</li>
  <li>Tipos arbitrarios.</li>
  <li>Puede contener duplicados.</li>
  <li>No tienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.</li>
  <li>Los elementos van ordenados por posición.</li>
  <li>Acceso a través de la sintaxis: <code>[index]</code></li>
  <li>Índices van de 0 a n-1, donde n es el número de elementos de la tupla.</li>
  <li>Son secuencias donde el orden de los elementos importa.</li>
  <li>Soportan anidamiento.</li>
</ul>

<h4>A diferencia de las listas:</h4>
<ul>
  <li>Se definen por medio de paréntesis.</li>
  <li>Inmutables.</li>
</ul>

<h4>¿Por qué tuplas?</h4>
<ul>
  <li>Representación de una colección fija de elementos (por ejemplo, una fecha).</li>
  <li>Pueden usarse en contextos que requieren inmutabilidad (por ejemplo, como claves de un diccionario).</li>
</ul>

<h3>3.6 Para terminar, volvemos a enfatizar un matiz importante…</h3>

<p><strong>En Python todo son objetos</strong></p>

<p>Cada objeto tiene:</p>
<ul>
  <li><strong>Identidad:</strong> Nunca cambia una vez creado. Es como la dirección de memoria. 
      El operador <code>is</code> compara identidad. La función <code>id()</code> devuelve identidad.</li>
  <li><strong>Tipo:</strong> Determina posibles valores y operaciones. La función <code>type()</code> 
      devuelve el tipo. No cambia.</li>
  <li><strong>Valor:</strong> Que pueden ser mutables e inmutables.
    <ul>
      <li><strong>Tipos mutables:</strong> list, dictionary, set y tipos definidos por el usuario.</li>
      <li><strong>Tipos inmutables:</strong> int, float, bool, string y tuple.</li>
    </ul>
  </li>
</ul>

```python
# Asignación
list_numbers = [1, 2, 3] # Lista (mutable)
tuple_numbers = (1, 2, 3) # Tupla (inmutable)
print(list_numbers[0])
print(tuple_numbers[0])
list_numbers[0] = 100
# tuple_numbers[0] = 100
print(list_numbers)
print(tuple_numbers)
tuple_l_numbers = list(tuple_numbers)
tuple_l_numbers[0] = 100
tuple_numbers = tuple(tuple_l_numbers)
print(tuple_numbers)
```

```python
# Identidad
list_numbers = [1, 2, 3]
tuple_numbers = (1, 2, 3)
print('Id list_numbers: ' + str(id(list_numbers)))
print('Id tuple_numbers: ' + str(id(tuple_numbers)))
list_numbers += [4, 5, 6] # La lista original se extiende
tuple_numbers += (4, 5, 6) # Se crea un nuevo objeto
print(list_numbers)
print(tuple_numbers)
print('Id list_numbers: ' + str(id(list_numbers)))
print('Id tuple_numbers: ' + str(id(tuple_numbers)))
```

```python
# Referencias
list_numbers = [1, 2, 3]
list_numbers_2 = list_numbers # list_numbers_2 referencia a list_numbers
print('Id list_numbers: ' + str(id(list_numbers)))
print('Id list_numbers_2: ' + str(id(list_numbers_2)))
list_numbers.append(4) # Se actualiza list_numbers2 también
print(list_numbers)
print(list_numbers_2)
print('Id list_numbers: ' + str(id(list_numbers)))
print('Id list_numbers_2: ' + str(id(list_numbers_2)))
```

```python
text = "Hola" # Inmutable
text_2 = text # Referencia
print('Id text: ' + str(id(text)))
print('Id text_2: ' + str(id(text_2)))
text += " Mundo"
print(text)
print(text_2)
print('Id text: ' + str(id(text)))
print('Id text_2: ' + str(id(text_2)))
```

```python
teams = ["Team A", "Team B", "Team C"] # Mutable
player = (23, teams) # Inmutable
print(type(player))
print(player)
print(id(player))
teams[2] = "Team J"
print(player)
print(id(player))
player[1][0] = 'Team XX'
print(teams)
```

<h4>Bucle <code>while</code></h4>
<ul>
  <li>Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta que una condición evalue a <code>False</code>).</li>
  <li>Si la condición evalua a <code>False</code> desde el principio, el bloque de código nunca se ejecuta.</li>
  <li>Cuidado con los bucles infinitos.</li>
</ul>

<h5>Formato general:</h5>
<pre><code>while test:    # Mientras se cumple la condición
    statements # Instrucciones a ejecutar</code></pre>

```python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]
while indice < 3:
    print(numeros[indice])
    indice =indice+1
```

```python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
print('Contador:', contador)
print('Indice:', indice)
```

```python
nombre = 'Pablo'
while nombre: # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]
```

```python
# #bucle infinito
i = 0
while i < 10:
    print(i)
    i += 1
```

  <li><code>break</code> permite terminar el bucle por completo.</li>
  <li><code>continue</code> permite saltar a la siguiente iteración, continuando con el bucle.</li>
  <li>Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (<code>if</code>).</li>
</ul>

```python
# # Ejemplo: break
rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]
for rating in movie_ratings:
    print(rating)
    if rating == rating_to_find:
        print("Found")
        break
print(rating)
```

```python
for i in range(10):
    if i % 2 == 0:
        continue # Si el número es par, salta a la siguiente iteración
    print(f'Odd number {i}')
```

<ul>
  <li>Observa como la sentencia <code>continue</code> te puede ayudar a reducir el número de niveles de anidamiento.</li>
  <li>Sin <code>continue</code> el anterior ejemplo sería:</li>
</ul>

```python
for i in range(10):
    if i % 2 != 0:
        print('Numero impar:', end=' ')
        print(i)
```

```python
for i in range(5):
    for a in range(2):
        print(f"{i} - {a}")
        if i == 3 and a == 0:
            break
```

<h4>Else</h4>
<ul>
  <li>Los bucles pueden tener una sentencia <code>else</code>.</li>
  <li>Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros lenguajes.</li>
  <li>Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de un <code>break</code>.</li>
</ul>

```python
lista = [60,60,30,60]
element_to_find = 60
for element in lista:
    if element == element_to_find:
        print("found")
        break
else:
    print("not found")
```

```python
# with flag
lista = [10,30,50,40]
element_to_find = 30
found = False
for element in lista:
    if element == element_to_find:
        found = True
        break
if found:
    print("dato encontrado")
else:
    print("not found")
```

<h2>4. List Comprehensions</h2>

<h4>List Comprehensions</h4>
<ul>
  <li>Permiten construir listas a través de la ejecución repetida (sentencia <code>for</code>) de una expresión para cada item de un objeto iterable.</li>
  <li>Van entre ‘<code>[</code>’ y ‘<code>]</code>’. Esto es indicativo de que estamos construyendo una lista.</li>
</ul>

<h5>Sintaxis:</h5>
<pre><code>[&lt;expression&gt; for &lt;item&gt; in &lt;iterable&gt;]</code></pre>

<h5>Ejemplo: repetir los carácteres de un string.</h5>

```python
[char*3 for char in "Enrique"]
```

```python
lista = []
for char in 'Enrique':
    lista.append(char)
print(lista)
```

<ul>
  <li>Comúnmente, el <code>item</code> de la sentencia <code>for</code> aparecerá en la expresión principal, pero eso no es obligatorio.</li>
</ul>

```python
[2 for _ in 'Enrique']
```

<h5>4.0.1 Versión extendida</h5>
<ul>
  <li>Se puede especificar un filtro (sentencia <code>if</code>) para obtener únicamente los elementos que cumplan cierta condición.</li>
</ul>

<h6>Sintaxis:</h6>
<pre><code>[&lt;expression&gt; for &lt;item&gt; in &lt;iterable&gt; if &lt;condition&gt;]</code></pre>

<h6>Ejemplo: obtener números pares:</h6>

```python
[y for y in range(9) if y % 2 == 0]
```

<ul>
  <li>Las list comprehension no son realmente requeridas, ya que siempre podemos escribir un bucle equivalente.</li>
</ul>

```python
pares = []
for x in range(9):
    if x % 2 == 0:
        pares.append(x)
print(pares)
```

<h5>4.0.2 Versión completa</h5>
<ul>
  <li>La sentencia <code>if</code> de una list comprehension también puede contener una expresión alternativa.</li>
</ul>

<h6>Sintaxis:</h6>
<pre><code>[&lt;expression_1&gt; if &lt;condition&gt; else &lt;expression_2&gt; for &lt;item&gt; in &lt;iterable&gt;]</code></pre>

<h6>Ejemplo: poner a cero los números pares.</h6>

```python
[0 if x % 2 == 0 else x for x in range(9)]
```

```python
[x**2 if x > 2 else x for x in range(-4,5) if x > 0]
```

```python
lista = []
for x in range(-4,5):
    if x > 0:
        if x > 2:
            lista.append(x**2)
        else:
            lista.append(x)
print(lista)
```

<h5>4.0.3 Anidamiento</h5>
<ul>
  <li>Las list comprehensions soportan anidamiento en sus expresiones.</li>
</ul>

<h6>Ejemplo: bucle anidado para obtener una lista de tuplas que combinan los elementos de dos listas dadas.</h6>

```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5]
[(x, y) for x in lista_1 for y in lista_2]
```

<h5>4.0.4 Pros y contras</h5>
<ul>
  <li>Principales ventajas de las comprehensions en comparación a un bucle convencional:
    <ul>
      <li>Expresión compacta y legible, si estás familiarizado con la sintaxis.</li>
      <li>Mejor rendimiento.</li>
    </ul>
  </li>
  <li>Desventaja: no escalan bien. Una list comprehension se puede convertir rápidamente en una expresión difícil de entender.</li>
</ul>

<h5>4.0.5 Otros tipos de comprehensions</h5>
<h6>Dictionary comprehensions</h6>
<ul>
  <li>Usando ‘<code>{</code>’ y ‘<code>}</code>’ como delimitadores y una expresión ‘<code>clave : valor</code>’, se obtiene un diccionario en lugar de una lista.</li>
</ul>
<h6>Ejemplo: diccionario donde cada valor es el cuadrado de la clave.</h6>

```python
d = {x : x*x for x in range(10)}
print(d)
print(type(d))
```

```python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
shopping = {k:v for k,v in zip(items,price)}
print(shopping)
```

```python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
for k in zip(items, price):
    print(type(k))
    print(k)
```

<h6>Set comprehensions</h6>
<ul>
  <li>Usando ‘<code>{</code>’ y ‘<code>}</code>’ como delimitadores y una expresión simple (al igual que en las list comprehensions), se obtiene un conjunto.</li>
</ul>
<h6>Ejemplo: conjunto que incluye los 10 primeros números naturales.</h6>

```python
c = {x for x in range(10)}
print(c)
print(type(c))
```

<h3>Tuple comprehensions</h3>

```python
tupla = tuple(x for x in range(4))
print(tupla)
```

<h3>5. Excepciones</h3>

<h4>Excepciones</h4>
<ul>
  <li>Las excepciones son eventos que representan situaciones excepcionales.</li>
  <li>Alteran el flujo de ejecución convencional.</li>
  <li>Python lanza excepciones automáticamente cuando se producen errores.</li>
  <li>El programador puede lanzar excepciones de manera explícita y también capturar excepciones para actuar como se crea conveniente.</li>
</ul>

<h5>Try/except</h5>
<ul>
  <li>Permite capturar excepciones y actuar en consecuencia.</li>
</ul>
<h6>Ejemplo: error de acceso fuera de rango.</h6>

```python
lista = [6, 1, 0, 5]
lista[4]
print('Código tras el error')
```

```python
lista = [6, 1, 0, 5]
i = 300
try:
    a = lista[i]
except IndexError:
    print('He capturado la excepción de tipo IndexError')
    a = 0
print('Código tras el bloque try')
print(a)
```

<ul>
  <li>Normalmente, al capturar excepciones, querremos ser lo más específicos posible, pero también se puedes usar una sentencia <code>try-except</code> que capture cualquier error.</li>
</ul>

```python
try:
    4/0
except:
    print('He capturado la división por cero')
```

<ul>
  <li>Recoger la excepción e imprimir el mensaje de error.</li>
</ul>

```python
lista = [6, 1, 0, 5]
try:
    print(lista[4])
except Exception as e:
    print("Error = " + str(e))
print('Reacheable Code')
```

<h5>Try/finally</h5>
<ul>
  <li>A través de <code>finally</code> podemos especificar código que queremos que se ejecute siempre (independientemente de si se produce la excepción o no).</li>
  <li>Se suele usar para liberar recursos.</li>
</ul>

```python
lista = [6, 1, 0, 5]
try:
    a = lista[1]
except IndexError:
    print('Exception IndexError Captured')
finally:
    print('Bloque finally')
    b = lista[2]
print('Código tras el bloque try')
print(b)
```

<h5>raise</h5>
<ul>
  <li>La sentencia <code>raise</code> nos permite lanzar excepciones de manera explícita.</li>
</ul>

```python
lista = [6, 1, 0, 5]
indice = 4
try:
    if indice >= len(lista):
        raise IndexError('Índice fuera de rango')
except IndexError:
    print('Excepción de tipo IndexError capturada')
```

<li>4. Escribe un programa que, dada una lista de palabras, compruebe si alguna empieza por ‘a’ y tiene más de 9 caracteres.
```python
palabras = ["apple", "abracadabra", "banana", "aventurero", "kiwi"]
encontrada = None
for palabra in palabras:
    if palabra.startswith('a') and len(palabra) > 9:
        encontrada = palabra
        break
if encontrada:
    print(f"Éxito: '{encontrada}' encontrada")
else:
    print("Fracaso: No se encontró ninguna palabra")
```

Por ejemplo, para el diccionario {‘a’: 4.3, ‘b’: 1, ‘c’: 7.8, ‘d’: -5} la respuesta sería ‘c’, dado que 7.8 es el valor más alto de los números 4.3, 1, 7.8 y -5.</li>

<li>7. Dada la lista a = [2, 4, 6, 8] y la lista b = [7, 11, 15, 22], escribe un programa que itere las listas a y b y multiplique cada elemento de a que sea mayor que 5 por cada elemento de b que sea menor que 14. El programa debe mostrar los resultados por pantalla.</li>

```python
a = [2, 4, 6, 8]
b = [7, 11, 15, 22]
for num_a in a:
    if num_a > 5:
        for num_b in b:
            if num_b < 14:
                print(f"{num_a} * {num_b} = {num_a * num_b}")
```
