*   **Referencias a objetos:**
    *   Las variables y los objetos se almacenan en diferentes zonas de memoria.

    *   Objetos sí que pueden referenciar a otros objetos. Ejemplo: listas.

*   **Sentencia de asignación:**
    <!-- FORMATO A SEGUIR: <nombre_variable> '=' <objeto> -->

```python
 # # Asignación de variables
a = 5
print(a)
```

```python
a = 1 # entero
b = 4.0 # coma flotante
c = "ITQ" # string
d = 10 + 1j # numero complejo
e = True #False # boolean
f = None # None
# visualizar valor de las variables y su tipo
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
print(d)
print(type(d))
print(e)
print(type(e))
print(f)
print(type(f))
```

<h2>Identificadores</h2>

* Podemos obtener un identificador único para los objetos referenciados por variables.
* Este identificador se obtiene a partir de la dirección de memoria.

```python
a = 3
print(id(a))
a = 'Pablo García'
print(id(a))
a = 4.5
print(id(a))
```

* Referencias compartidas: un mismo objeto puede ser referenciado por más de una variable.
* Variables que referencian al mismo objeto tienen mismo identificador.


```python
a = 4567
print(id(a))
```

```python
b = a
print(id(b))
```

```python
c = 4567
print(id(c))
```

```python
a = 25
b = 25
print(id(a))
print(id(b))
print(id(25))
```

* Union, interseccion y diferencia

```python
set1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set2 = set([0, 1, 2, 3, 4, 42])
# union
print(set1 | set2)
# intersección
print(set1 & set2)
# diferencia
print(set1 - set2)
print(set2 - set1)
```

```python
a = [1,2]
b = [2,3]
print(set(a) & set(b))
```

```python
#Ademas de los operadores, que operan únicamente con Sets, también se pueden usar metodos que pueden operar sobre cualquier objeto iterable.
conjunto = {0, 1, 2}
lista = [1, 3, 3]
print(conjunto.union(lista))
print(conjunto.intersection(lista))
print(conjunto.difference(lista))
```

```python
set1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set2 = set([0, 1, 2, 3, 4])
print(set2.issubset(set1))
print(set1.issuperset(set2))
print(set1.isdisjoint(set2))
```

* Modificación de conjuntos

```python
# A través de operador de asignación
set1 = {'a', 'b', 'c'}
set2 = {'a', 'd'}
# set1 |= set2 # set1 = set1 | set2
# set1 &= set2
set1 -= set2
print(set1)
```

```python
# A través de método 'update'.
set1 = {'a', 'b', 'c'}
set2 = {'a', 'd'}
# set1.update(set2)
# set1.intersection_update(set2)
set1.difference_update(set2)
print(set1)
```

```python
# A través de métodos 'add' y 'remove'.
set1 = {'a', 'b', 'c'}
set1.add('d')
set1.remove('a')
print(set1)
```

<h2>3.4 Tuplas</h2>

<h4>Al igual que las listas:</h4>
<ul>
  <li>Colección de elementos.</li>
  <li>Tipos arbitrarios.</li>
  <li>Puede contener duplicados.</li>
  <li>No tienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.</li>
  <li>Los elementos van ordenados por posición.</li>
  <li>Acceso a través de la sintaxis: <code>[index]</code></li>
  <li>Índices van de 0 a n-1, donde n es el número de elementos de la tupla.</li>
  <li>Son secuencias donde el orden de los elementos importa.</li>
  <li>Soportan anidamiento.</li>
</ul>

<h4>A diferencia de las listas:</h4>
<ul>
  <li>Se definen por medio de paréntesis.</li>
  <li>Inmutables.</li>
</ul>

<h4>¿Por qué tuplas?</h4>
<ul>
  <li>Representación de una colección fija de elementos (por ejemplo, una fecha).</li>
  <li>Pueden usarse en contextos que requieren inmutabilidad (por ejemplo, como claves de un diccionario).</li>
</ul>

<h3>3.6 Para terminar, volvemos a enfatizar un matiz importante…</h3>

<p><strong>En Python todo son objetos</strong></p>

<p>Cada objeto tiene:</p>
<ul>
  <li><strong>Identidad:</strong> Nunca cambia una vez creado. Es como la dirección de memoria. 
      El operador <code>is</code> compara identidad. La función <code>id()</code> devuelve identidad.</li>
  <li><strong>Tipo:</strong> Determina posibles valores y operaciones. La función <code>type()</code> 
      devuelve el tipo. No cambia.</li>
  <li><strong>Valor:</strong> Que pueden ser mutables e inmutables.
    <ul>
      <li><strong>Tipos mutables:</strong> list, dictionary, set y tipos definidos por el usuario.</li>
      <li><strong>Tipos inmutables:</strong> int, float, bool, string y tuple.</li>
    </ul>
  </li>
</ul>

```python
# Asignación
list_numbers = [1, 2, 3] # Lista (mutable)
tuple_numbers = (1, 2, 3) # Tupla (inmutable)
print(list_numbers[0])
print(tuple_numbers[0])
list_numbers[0] = 100
# tuple_numbers[0] = 100
print(list_numbers)
print(tuple_numbers)
tuple_l_numbers = list(tuple_numbers)
tuple_l_numbers[0] = 100
tuple_numbers = tuple(tuple_l_numbers)
print(tuple_numbers)
```

```python
# Identidad
list_numbers = [1, 2, 3]
tuple_numbers = (1, 2, 3)
print('Id list_numbers: ' + str(id(list_numbers)))
print('Id tuple_numbers: ' + str(id(tuple_numbers)))
list_numbers += [4, 5, 6] # La lista original se extiende
tuple_numbers += (4, 5, 6) # Se crea un nuevo objeto
print(list_numbers)
print(tuple_numbers)
print('Id list_numbers: ' + str(id(list_numbers)))
print('Id tuple_numbers: ' + str(id(tuple_numbers)))
```

```python
# Referencias
list_numbers = [1, 2, 3]
list_numbers_2 = list_numbers # list_numbers_2 referencia a list_numbers
print('Id list_numbers: ' + str(id(list_numbers)))
print('Id list_numbers_2: ' + str(id(list_numbers_2)))
list_numbers.append(4) # Se actualiza list_numbers2 también
print(list_numbers)
print(list_numbers_2)
print('Id list_numbers: ' + str(id(list_numbers)))
print('Id list_numbers_2: ' + str(id(list_numbers_2)))
```

```python
text = "Hola" # Inmutable
text_2 = text # Referencia
print('Id text: ' + str(id(text)))
print('Id text_2: ' + str(id(text_2)))
text += " Mundo"
print(text)
print(text_2)
print('Id text: ' + str(id(text)))
print('Id text_2: ' + str(id(text_2)))
```

```python
teams = ["Team A", "Team B", "Team C"] # Mutable
player = (23, teams) # Inmutable
print(type(player))
print(player)
print(id(player))
teams[2] = "Team J"
print(player)
print(id(player))
player[1][0] = 'Team XX'
print(teams)
```

<h4>Bucle <code>while</code></h4>
<ul>
  <li>Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta que una condición evalue a <code>False</code>).</li>
  <li>Si la condición evalua a <code>False</code> desde el principio, el bloque de código nunca se ejecuta.</li>
  <li>Cuidado con los bucles infinitos.</li>
</ul>

<h5>Formato general:</h5>
<pre><code>while test:    # Mientras se cumple la condición
    statements # Instrucciones a ejecutar</code></pre>

```python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]
while indice < 3:
    print(numeros[indice])
    indice =indice+1
```

```python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
print('Contador:', contador)
print('Indice:', indice)
```

```python
nombre = 'Pablo'
while nombre: # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]
```

```python
# #bucle infinito
i = 0
while i < 10:
    print(i)
    i += 1
```

  <li><code>break</code> permite terminar el bucle por completo.</li>
  <li><code>continue</code> permite saltar a la siguiente iteración, continuando con el bucle.</li>
  <li>Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (<code>if</code>).</li>
</ul>

```python
# # Ejemplo: break
rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]
for rating in movie_ratings:
    print(rating)
    if rating == rating_to_find:
        print("Found")
        break
print(rating)
```

```python
for i in range(10):
    if i % 2 == 0:
        continue # Si el número es par, salta a la siguiente iteración
    print(f'Odd number {i}')
```

<ul>
  <li>Observa como la sentencia <code>continue</code> te puede ayudar a reducir el número de niveles de anidamiento.</li>
  <li>Sin <code>continue</code> el anterior ejemplo sería:</li>
</ul>

```python
for i in range(10):
    if i % 2 != 0:
        print('Numero impar:', end=' ')
        print(i)
```

```python
for i in range(5):
    for a in range(2):
        print(f"{i} - {a}")
        if i == 3 and a == 0:
            break
```

<h4>Else</h4>

  <li>Los bucles pueden tener una sentencia <code>else</code>.</li>
  <li>Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros lenguajes.</li>
  <li>Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de un <code>break</code>.</li>
</ul>

```python
lista = [60,60,30,60]
element_to_find = 60
for element in lista:
    if element == element_to_find:
        print("found")
        break
else:
    print("not found")
```

```python
# with flag
lista = [10,30,50,40]
element_to_find = 30
found = False
for element in lista:
    if element == element_to_find:
        found = True
        break
if found:
    print("dato encontrado")
else:
    print("not found")
```

<h2>4. List Comprehensions</h2>

<h4>List Comprehensions</h4>
<ul>
  <li>Permiten construir listas a través de la ejecución repetida (sentencia <code>for</code>) de una expresión para cada item de un objeto iterable.</li>
  <li>Van entre ‘<code>[</code>’ y ‘<code>]</code>’. Esto es indicativo de que estamos construyendo una lista.</li>
</ul>

<h5>Sintaxis:</h5>
<pre><code>[&lt;expression&gt; for &lt;item&gt; in &lt;iterable&gt;]</code></pre>

<h5>Ejemplo: repetir los carácteres de un string.</h5>

```python
[char*3 for char in "Enrique"]
```

```python
lista = []
for char in 'Enrique':
    lista.append(char)
print(lista)
```

<ul>
  <li>Comúnmente, el <code>item</code> de la sentencia <code>for</code> aparecerá en la expresión principal, pero eso no es obligatorio.</li>
</ul>

```python
[2 for _ in 'Enrique']
```

<h5>4.0.1 Versión extendida</h5>
<ul>
  <li>Se puede especificar un filtro (sentencia <code>if</code>) para obtener únicamente los elementos que cumplan cierta condición.</li>
</ul>

<h6>Sintaxis:</h6>
<pre><code>[&lt;expression&gt; for &lt;item&gt; in &lt;iterable&gt; if &lt;condition&gt;]</code></pre>

<h6>Ejemplo: obtener números pares:</h6>

```python
[y for y in range(9) if y % 2 == 0]
```

<ul>
  <li>Las list comprehension no son realmente requeridas, ya que siempre podemos escribir un bucle equivalente.</li>
</ul>

```python
pares = []
for x in range(9):
    if x % 2 == 0:
        pares.append(x)
print(pares)
```

<h5>4.0.2 Versión completa</h5>
<ul>
  <li>La sentencia <code>if</code> de una list comprehension también puede contener una expresión alternativa.</li>
</ul>

<h6>Sintaxis:</h6>
<pre><code>[&lt;expression_1&gt; if &lt;condition&gt; else &lt;expression_2&gt; for &lt;item&gt; in &lt;iterable&gt;]</code></pre>

<h6>Ejemplo: poner a cero los números pares.</h6>

```python
[0 if x % 2 == 0 else x for x in range(9)]
```

```python
[x**2 if x > 2 else x for x in range(-4,5) if x > 0]
```

```python
lista = []
for x in range(-4,5):
    if x > 0:
        if x > 2:
            lista.append(x**2)
        else:
            lista.append(x)
print(lista)
```

<h5>4.0.3 Anidamiento</h5>
<ul>
  <li>Las list comprehensions soportan anidamiento en sus expresiones.</li>
</ul>

<h6>Ejemplo: bucle anidado para obtener una lista de tuplas que combinan los elementos de dos listas dadas.</h6>

```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5]
[(x, y) for x in lista_1 for y in lista_2]
```

<h5>4.0.4 Pros y contras</h5>
<ul>
  <li>Principales ventajas de las comprehensions en comparación a un bucle convencional:
    <ul>
      <li>Expresión compacta y legible, si estás familiarizado con la sintaxis.</li>
      <li>Mejor rendimiento.</li>
    </ul>
  </li>
  <li>Desventaja: no escalan bien. Una list comprehension se puede convertir rápidamente en una expresión difícil de entender.</li>
</ul>

<h5>4.0.5 Otros tipos de comprehensions</h5>
<h6>Dictionary comprehensions</h6>
<ul>
  <li>Usando ‘<code>{</code>’ y ‘<code>}</code>’ como delimitadores y una expresión ‘<code>clave : valor</code>’, se obtiene un diccionario en lugar de una lista.</li>
</ul>
<h6>Ejemplo: diccionario donde cada valor es el cuadrado de la clave.</h6>

<h3>5. Excepciones</h3>

<h4>Excepciones</h4>
<ul>
  <li>Las excepciones son eventos que representan situaciones excepcionales.</li>
  <li>Alteran el flujo de ejecución convencional.</li>
  <li>Python lanza excepciones automáticamente cuando se producen errores.</li>
  <li>El programador puede lanzar excepciones de manera explícita y también capturar excepciones para actuar como se crea conveniente.</li>
</ul>

<h5>Try/except</h5>
<ul>
  <li>Permite capturar excepciones y actuar en consecuencia.</li>
</ul>
<h6>Ejemplo: error de acceso fuera de rango.</h6>

```python
lista = [6, 1, 0, 5]
lista[4]
print('Código tras el error')
```

```python
lista = [6, 1, 0, 5]
i = 300
try:
    a = lista[i]
except IndexError:
    print('He capturado la excepción de tipo IndexError')
    a = 0
print('Código tras el bloque try')
print(a)
```

<ul>
  <li>Normalmente, al capturar excepciones, querremos ser lo más específicos posible, pero también se puedes usar una sentencia <code>try-except</code> que capture cualquier error.</li>
</ul>

```python
try:
    4/0
except:
    print('He capturado la división por cero')
```

<ul>
  <li>Recoger la excepción e imprimir el mensaje de error.</li>
</ul>

```python
lista = [6, 1, 0, 5]
try:
    print(lista[4])
except Exception as e:
    print("Error = " + str(e))
print('Reacheable Code')
```

<h5>Try/finally</h5>
<ul>
  <li>A través de <code>finally</code> podemos especificar código que queremos que se ejecute siempre (independientemente de si se produce la excepción o no).</li>
  <li>Se suele usar para liberar recursos.</li>
</ul>

```python
lista = [6, 1, 0, 5]
try:
    a = lista[1]
except IndexError:
    print('Exception IndexError Captured')
finally:
    print('Bloque finally')
    b = lista[2]
print('Código tras el bloque try')
print(b)
```

<h5>raise</h5>
<ul>
  <li>La sentencia <code>raise</code> nos permite lanzar excepciones de manera explícita.</li>
</ul>

```python
lista = [6, 1, 0, 5]
indice = 4
try:
    if indice >= len(lista):
        raise IndexError('Índice fuera de rango')
except IndexError:
    print('Excepción de tipo IndexError capturada')
```

<h3>1.8 Funciones Lambda</h3>
<ul>
  <li>Es posible definir funciones anónimas (sin nombre).</li>
  <li>Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia <code>def</code> no puede (por ejemplo, dentro de una lista o como parámetro de una función).</li>
  <li>Útiles cuando se quiere pasar una función como argumento a otra.</li>
</ul>

<p><strong>Formato general:</strong></p>
<pre><code>lambda &lt;lista de argumentos&gt; : &lt;valor a retornar&gt;</code></pre>

<ul>
  <li>Importante: <code>&lt;valor a retornar&gt;</code> no es un conjunto de instrucciones. Es simplemente una expresión <code>return</code> que omite esta palabra.</li>
  <li>Las funciones definidas con <code>def</code> son más generales:
    <ul>
      <li>Cualquier cosa que implementes en un lambda, lo puedes implementar como una función convencional con <code>def</code>, pero no viceversa.</li>
    </ul>
  </li>
</ul>

```python
# Ejemplo: función suma.
def suma(x, y, z):
    return x + y + z
print(suma(1, 2, 3))
# Ejemplo: función suma como expresión lambda.
suma2 = lambda x, y, z : x + y + z
print(suma2(1, 2, 3))
```

```python
# La funcion 'sort' puede recibir una función optional como parámetro.
# Esta función 'key' se invoca para cada elemento de la lista antes de realizar las comparaciones.
# Ejemplo: ordenar strings por número de carácteres.
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']
cities.sort()
print(cities)
cities.sort(key = lambda x : len(x))
print(cities)
cities.sort(key = len)
print(cities)
def longitud(x):
    return len(x)
cities.sort(key = longitud)
print(cities)
```

```python
# Mismo ejemplo sin lambda
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']
def num_caracteres(x):
    return len(x)
cities.sort(key = num_caracteres)
print(cities)
```

<ul>
  <li>Diccionarios se pueden utilizar como ‘switch’ en otros lenguajes.</li>
  <li>Ideal para escoger entre varias opciones.</li>
</ul>

```python
def switch_dict(operator, x, y):
    # crea un diccionario de opciones y selecciona 'operator'
    return {
        'add': lambda: x + y,
        'sub': lambda: x - y,
        'mul': lambda: x * y,
        'div': lambda: x / y,
    }.get(operator, lambda: None)() # retorna None si no encuentra 'operator'
print(switch_dict('add',2,2))
print(switch_dict('div',10,2))
print(switch_dict('mod',17,3))
```

<h3>1.9 Scopes</h3>
<ul>
  <li>La localización de una asignación a una variable en el código determina desde donde puedes acceder a esa variable.</li>
  <li>Por ejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.</li>
  <li>El alcance de la visibilidad de una variable determina su <em>scope</em>.</li>
  <li>El término <em>scope</em> hace referencia a un espacio de nombres (<em>namespace</em>). Por ejemplo, una función establece su propio <em>namespace</em>.</li>
</ul>

```python
X = 10
def func():
    X = 20 # Local a la función. Es una variable diferente, no visible fuera de 'func'
    def func_int():
        X = 30
        print(X)
    func_int()
func()
```

<h4>Regla LEGB</h4>
<p>Dada una función F, tenemos los siguientes scopes:</p>
<ul>
  <li><strong>Local:</strong> variables locales a F.</li>
  <li><strong>Enclosing:</strong> variables localizadas en funciones que contienen a (o por encima de) F.</li>
  <li><strong>Global:</strong> variables definidas en el módulo (fichero) que no están contenidas en ninguna función. Este scope no abarca más de un fichero.</li>
  <li><strong>Built-ins:</strong> proporcionadas por el lenguaje.</li>
</ul>
<p>La resolución de nombres ocurre de abajo a arriba.</p>

```python
# Global (X y func)
X = 99
def func(Y): # Local (Y y Z)
    Z = X + Y
    return Z
print(func(1))
```

```python
X = 1
def func():
    X = 2 # X es una variable diferente
func()
print(X)
```

<ul>
  <li>La sentencia <code>global</code> nos permite modificar una variable global desde dentro de una función.</li>
</ul>

```python
X = 1
def func():
    global X
    X = 2
func()
print(X)
```

<ul>
  <li>No hay necesidad de usar <code>global</code> para referenciar variables. Sólo es necesario para modificarlas.</li>
</ul>

```python
y, z = 1, 2
def todas_globales():
    global x
    x = y + z # Las tres variables son globales.
todas_globales()
print(x)
print(y)
print(z)
```

```python
X = 77
def f1():
    X = 88 # Enclosing scope (para f2)
    def f2():
        print(X)
    f2()
f1()
```

<h4>Closures</h4>
<ul>
  <li>Las funciones pueden recordar su <em>enclosing scope</em>, independientemente de si éstos continúan existiendo o no.</li>
</ul>

```python
def f1():
    X = 88 # Enclosing scope (para f2)
    def f2():
        print(X)
    return f2 # Devuelve f2, sin invocarla
accion = f1()
accion()
 
```

<ul>
  <li>Sentencia <code>nonlocal</code> para modificar variables que no son locales, pero tampoco globales.</li>
</ul>

```python
def f1():
    contador = 10
    def f2():
        nonlocal contador
        contador += 1
        print(contador)
    return f2
accion = f1()
accion()
accion()
accion()
```

<h4>8.0.1 ¿Por qué molestarse con .npy?</h4>
<ul>
  <li>Bases de datos suelen estar en CSV o txt, que requieren lectura con streams (parsing).</li>
  <li>¿Para qué escribir los datos de nuevo?</li>
</ul>

```python
import os
import numpy as np

# 1. Crear carpeta si no existe
carpeta = "res"
os.makedirs(carpeta, exist_ok=True)

# 2. Ruta del archivo CSV
ruta_csv = os.path.join(carpeta, "fdata.csv")

# 3. Si el CSV no existe, crearlo con formato limpio
if not os.path.exists(ruta_csv):
    print(f"Creando {ruta_csv}...")
    datos = np.random.randint(0, 10, size=(1000, 1000))
    # Guardar sin espacios y con delimitador simple
    np.savetxt(ruta_csv, datos, delimiter=',', fmt='%d', newline='\n')
    print(f"✅ CSV creado: {ruta_csv}")

# 4. Leer y procesar el CSV CORRECTAMENTE
print(f"Leyendo {ruta_csv}...")

# Opción A: Usar numpy para cargar directamente (RECOMENDADO)
try:
    data_array = np.loadtxt(ruta_csv, delimiter=',', dtype=np.int8)
    print("✅ Cargado con np.loadtxt")
except Exception as e:
    print(f"Error con loadtxt: {e}")
    
    # Opción B: Procesamiento manual robusto
    with open(ruta_csv, 'r') as f:
        lineas = f.readlines()
    
    # Procesar cada línea
    datos_procesados = []
    for linea in lineas:
        # Limpiar: eliminar espacios, saltos de línea, dividir por coma
        numeros = linea.strip().replace(' ', '').split(',')
        # Filtrar elementos vacíos
        numeros = [n for n in numeros if n]
        datos_procesados.extend(numeros)
    
    # Convertir a enteros y reshape
    datos_procesados = [int(n) for n in datos_procesados if n.strip()]
    
    # Asegurar tamaño correcto
    if len(datos_procesados) < 1000000:
        datos_procesados.extend([0] * (1000000 - len(datos_procesados)))
    datos_procesados = datos_procesados[:1000000]
    
    data_array = np.array(datos_procesados, dtype=np.int8).reshape(1000, 1000)

# 5. Mostrar información
print(f"📊 Array shape: {data_array.shape}")
print(f"📊 Tipo: {data_array.dtype}")
print("Muestra (3x3):")
print(data_array[:3, :3])

# 6. Guardar como .npy
ruta_npy = os.path.join(carpeta, "o_fdata.npy")
np.save(ruta_npy, data_array)
print(f"💾 Guardado como: {ruta_npy}")

# 7. Verificar que se guardó
if os.path.exists(ruta_npy):
    cargado = np.load(ruta_npy)
    print(f"✅ Verificado: {cargado.shape}, iguales: {np.array_equal(data_array, cargado)}")
```

<h3>4. Cómo encontrar el tamaño de memoria de cualquier array</h3>

```python
import numpy as np

# Crear un array de ejemplo
Z = np.zeros((10, 10))
print("Array de ejemplo Z (10x10):")
print(Z)

# Calcular tamaño de memoria
tamaño_bytes = Z.nbytes
tamaño_kb = tamaño_bytes / 1024

print(f"\n Tamaño de memoria:")
print(f"Bytes: {tamaño_bytes:,} B")
print(f"Kilobytes: {tamaño_kb:.2f} KB")
print(f"Elementos: {Z.size}")
print(f"Tamaño por elemento: {Z.itemsize} bytes")

# Función genérica
def tamaño_memoria(array):
    """Devuelve el tamaño de memoria de un array en formato legible"""
    bytes = array.nbytes
    if bytes < 1024:
        return f"{bytes} bytes"
    elif bytes < 1024**2:
        return f"{bytes/1024:.2f} KB"
    else:
        return f"{bytes/(1024**2):.2f} MB"

print(f"\n Usando función: {tamaño_memoria(Z)}")
```

<h3>16. ¿Cómo agregar un borde (relleno con 0's) alrededor de un array existente?</h3>

```python
# Solución ejercicio 16
import numpy as np

# Array existente
Z = np.ones((3, 3), dtype=int)
print("Array original 3x3:")
print(Z)

# Método 1: np.pad() (recomendado - más flexible)
Z_padded = np.pad(Z, pad_width=1, mode='constant', constant_values=0)
print("\nMétodo 1 - np.pad() con borde de 0's:")
print(Z_padded)
print(f"Nueva forma: {Z_padded.shape}")

# Método 2: Crear matriz más grande y copiar
print("\nMétodo 2 - Crear matriz más grande:")
rows, cols = Z.shape
Z_padded2 = np.zeros((rows+2, cols+2), dtype=Z.dtype)
Z_padded2[1:-1, 1:-1] = Z
print(Z_padded2)

# Método 3: Con catenación
print("\nMétodo 3 - Con catenación:")
zeros_row = np.zeros((1, cols), dtype=Z.dtype)
zeros_col = np.zeros((rows+2, 1), dtype=Z.dtype)

temp = np.vstack([zeros_row, Z, zeros_row])
Z_padded3 = np.hstack([zeros_col, temp, zeros_col])
print(Z_padded3)

# Ejemplo con borde más grueso
print("\n Ejemplo con borde de 2 píxeles:")
Z_thick_border = np.pad(Z, pad_width=2, mode='constant', constant_values=0)
print(Z_thick_border)
print(f"Forma original: {Z.shape} → Con borde: {Z_thick_border.shape}")

# Verificación
print(f"\n Verificación:")
print(f"¿Todos los métodos dan igual? {np.array_equal(Z_padded, Z_padded2) and np.array_equal(Z_padded2, Z_padded3)}")
print(f"¿Borde externo son 0's? {np.all(Z_padded[0,:]==0) and np.all(Z_padded[-1,:]==0) and np.all(Z_padded[:,0]==0) and np.all(Z_padded[:,-1]==0)}")
print(f"¿Interior preservado? {np.array_equal(Z_padded[1:-1, 1:-1], Z)}")
```

<h3>1.1 Ficheros</h3>
<ul>
  <li><strong>File</strong> es un tipo de objeto predefinido en Python (built-in).</li>
  <li>Permite acceder a ficheros desde programas Python.</li>
  <li>Los ficheros son de un tipo especial:
    <ul>
      <li>Son built-in, pero no son ni números, ni secuencias, ni mappings. Tampoco responden a operadores en expresiones.</li>
    </ul>
  </li>
  <li>La función <code>open</code> permite crear objetos de tipo fichero.</li>
</ul>

<h4>Formato general para abrir un fichero:</h4>
<pre><code>afile = open(filename, mode)</code></pre>

<ul>
  <li><code>mode</code> es opcional. Por defecto, los ficheros se abren en modo lectura.</li>
  <li>Los datos leídos de un fichero siempre se obtienen en formato string. Lo mismo ocurre con escritura.</li>
  <li>Los ficheros se deben cerrar invocando <code>close</code> (liberación de recursos).</li>
</ul>

```python
# Lectura desde fichero usando método 'read'. Devuelve todo el contenido del fichero.
mi_fichero = open('res/multiple_lines.txt')
print(mi_fichero.read())
mi_fichero.close()
```

```python
# Lectura linea a linea a través del bucle 'for'.
mi_fichero = open('res/multiple_lines.txt')
for linea in mi_fichero:
    print(f"{linea}")
mi_fichero.close()
```

```python
# leer de un archivo a una lista.
mi_fichero = open('res/multiple_lines.txt')
lineas = mi_fichero.readlines()
mi_fichero.close()
print(lineas)
```

<h4>1. Escribe una función que reciba una ruta de un fichero de texto y una cadena de caracteres a buscar y determine si la cadena aparece en el fichero.</h4>

```python
def buscar_cadena_en_archivo(ruta_archivo, cadena_buscar):
    """
    Determina si una cadena aparece en un archivo de texto.
    
    Parámetros:
    -----------
    ruta_archivo : str
        Ruta completa al archivo de texto
    cadena_buscar : str
        Cadena a buscar en el archivo
        
    Retorna:
    --------
    bool
        True si la cadena se encuentra en el archivo, False en caso contrario
    """
    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            contenido = archivo.read()
            return cadena_buscar in contenido
    except FileNotFoundError:
        print(f"Error: No se encontró el archivo '{ruta_archivo}'")
        return False
    except Exception as e:
        print(f"Error al leer el archivo: {e}")
        return False

# Ejemplo de uso
if __name__ == "__main__":
    # Crear un archivo de prueba
    texto_prueba = """Python es un lenguaje de programación interpretado.
Su filosofía hace hincapié en la legibilidad del código.
Es multiparadigma y multipropósito."""
    
    with open('ejemplo.txt', 'w', encoding='utf-8') as f:
        f.write(texto_prueba)
    
    # Probar la función
    resultado = buscar_cadena_en_archivo('ejemplo.txt', 'Python')
    print(f"¿'Python' está en el archivo? {resultado}")
    
    resultado2 = buscar_cadena_en_archivo('ejemplo.txt', 'Java')
    print(f"¿'Java' está en el archivo? {resultado2}")
```

<h4>2. Escribe una función que reciba una lista, una ruta destino y un número n. La función debe crear un fichero en la ruta especificada. El contenido del fichero serán los primeros n elementos de la lista. La función debe controlar de manera apropiada los posibles valores de n que estén fuera de rango.</h4>

```python
def escribir_primeros_n_elementos(lista, ruta_destino, n):
    """
    Escribe los primeros n elementos de una lista en un archivo.
    
    Parámetros:
    -----------
    lista : list
        Lista de elementos a escribir
    ruta_destino : str
        Ruta donde se creará el archivo
    n : int
        Número de elementos a escribir
        
    Retorna:
    --------
    bool
        True si la operación fue exitosa, False en caso contrario
    """
    try:
        # Validar que n sea un número entero positivo
        if not isinstance(n, int):
            print(f"Error: n debe ser un número entero, se recibió {type(n).__name__}")
            return False
            
        # Controlar valores de n fuera de rango
        if n <= 0:
            print(f"Advertencia: n={n} no es válido. Se escribirá archivo vacío.")
            elementos_a_escribir = []
        elif n > len(lista):
            print(f"Advertencia: n={n} es mayor que el tamaño de la lista ({len(lista)}). Se escribirán todos los elementos.")
            elementos_a_escribir = lista
        else:
            elementos_a_escribir = lista[:n]
        
        # Escribir en el archivo
        with open(ruta_destino, 'w', encoding='utf-8') as archivo:
            for elemento in elementos_a_escribir:
                archivo.write(f"{elemento}\n")
        
        print(f"✓ Archivo creado en: {ruta_destino}")
        print(f"  Elementos escritos: {len(elementos_a_escribir)} de {len(lista)}")
        return True
        
    except Exception as e:
        print(f" Error al crear el archivo: {e}")
        return False

# Ejemplo de uso
if __name__ == "__main__":
    # Lista de ejemplo
    frutas = ['manzana', 'banana', 'naranja', 'uva', 'pera', 'sandía', 'melón']
    
    print("Prueba 1: n dentro del rango")
    escribir_primeros_n_elementos(frutas, 'frutas_3.txt', 3)
    
    print("\nPrueba 2: n mayor que tamaño de lista")
    escribir_primeros_n_elementos(frutas, 'frutas_10.txt', 10)
    
    print("\nPrueba 3: n negativo")
    escribir_primeros_n_elementos(frutas, 'frutas_negativo.txt', -2)
    
    print("\nPrueba 4: n = 0")
    escribir_primeros_n_elementos(frutas, 'frutas_0.txt', 0)
    
    # Leer y mostrar contenido de uno de los archivos
    print("\nContenido de 'frutas_3.txt':")
    with open('frutas_3.txt', 'r', encoding='utf-8') as f:
        print(f.read())
```

<h4>3. Escribe una función que reciba una ruta de un fichero de texto devuelva un diccionario con la frecuencia de aparición de cada palabra. Ejemplo: un fichero que contenga la frase ‘es mejor que venga que que no venga’ devolverá el siguiente diccionario: {‘es’ : 1, ‘mejor’ : 1, ‘que’ : 3, ‘venga’ : 2, ‘no’ : 1}. Para dividir un string en palabras puedes hacer uso del método split.</h4>

```python
def frecuencia_palabras(ruta_archivo):
    """
    Calcula la frecuencia de aparición de cada palabra en un archivo de texto.
    
    Parámetros:
    -----------
    ruta_archivo : str
        Ruta al archivo de texto
        
    Retorna:
    --------
    dict
        Diccionario con palabras como claves y frecuencias como valores
        Retorna diccionario vacío si hay error
    """
    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            contenido = archivo.read()
        
        # Limpiar y dividir el contenido en palabras
        # Eliminar signos de puntuación y convertir a minúsculas
        import string
        contenido_limpio = contenido.translate(
            str.maketrans('', '', string.punctuation)
        ).lower()
        
        palabras = contenido_limpio.split()
        
        # Contar frecuencia de cada palabra
        frecuencia = {}
        for palabra in palabras:
            if palabra in frecuencia:
                frecuencia[palabra] += 1
            else:
                frecuencia[palabra] = 1
        
        return frecuencia
        
    except FileNotFoundError:
        print(f"Error: No se encontró el archivo '{ruta_archivo}'")
        return {}
    except Exception as e:
        print(f"Error al procesar el archivo: {e}")
        return {}

# Versión mejorada con opciones de procesamiento
def frecuencia_palabras_avanzada(ruta_archivo, ignorar_case=True, remover_puntuacion=True):
    """
    Calcula frecuencia de palabras con opciones avanzadas.
    
    Parámetros:
    -----------
    ruta_archivo : str
        Ruta al archivo de texto
    ignorar_case : bool
        Si True, convierte todo a minúsculas (default: True)
    remover_puntuacion : bool
        Si True, elimina signos de puntuación (default: True)
        
    Retorna:
    --------
    dict
        Diccionario de frecuencias
    """
    try:
        with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
            contenido = archivo.read()
        
        # Procesamiento según opciones
        if ignorar_case:
            contenido = contenido.lower()
        
        if remover_puntuacion:
            import string
            contenido = contenido.translate(
                str.maketrans('', '', string.punctuation)
            )
        
        palabras = contenido.split()
        
        # Usar dict comprehension o collections.Counter
        frecuencia = {}
        for palabra in palabras:
            frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
        
        return frecuencia
        
    except Exception as e:
        print(f"Error: {e}")
        return {}

# Ejemplo de uso
if __name__ == "__main__":
    # Crear archivo de prueba con el ejemplo dado
    texto_ejemplo = "es mejor que venga que que no venga"
    with open('ejemplo_frecuencia.txt', 'w', encoding='utf-8') as f:
        f.write(texto_ejemplo)
    
    print(" Ejemplo básico:")
    resultado = frecuencia_palabras('ejemplo_frecuencia.txt')
    print(f"Resultado: {resultado}")
    
    # Verificar que coincida con el ejemplo
    esperado = {'es': 1, 'mejor': 1, 'que': 3, 'venga': 2, 'no': 1}
    print(f"¿Coincide con lo esperado? {resultado == esperado}")
    
    # Ejemplo más complejo
    print("\n Ejemplo con texto más largo:")
    texto_largo = """Python es un lenguaje de programación interpretado.
    Python es multiparadigma y muy popular.
    Python, python, PYTHON... siempre Python."""
    
    with open('python.txt', 'w', encoding='utf-8') as f:
        f.write(texto_largo)
    
    resultado2 = frecuencia_palabras('python.txt')
    print("Frecuencias:")
    for palabra, freq in sorted(resultado2.items(), key=lambda x: x[1], reverse=True):
        print(f"  '{palabra}': {freq}")
    
    # Mostrar estadísticas
    print(f"\n Estadísticas:")
    print(f"Total de palabras distintas: {len(resultado2)}")
    print(f"Total de palabras: {sum(resultado2.values())}")
    print(f"Palabra más frecuente: {max(resultado2, key=resultado2.get)}")
```
