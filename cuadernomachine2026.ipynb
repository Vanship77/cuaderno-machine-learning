<h2>2.8 REPL (Read-Eval-Print Loop)</h2>

<p>
  Sistema interactivo para comunicarse con el ordenador en un lenguaje, en este caso Python. 
  Se debe cumplir el ciclo:
</p>
<ul>
  <li><strong>Read (Leer):</strong> El ordenador pueda leer comandos/instrucciones como entrada.</li>
  <li><strong>Evaluate (Evaluar):</strong> El código pueda ser procesado e interpretado.</li>
  <li><strong>Print (Imprimir):</strong> Los resultados puedan verse en la salida.</li>
  <li><strong>Loop (Bucle):</strong> Continuar con la conversación en un ciclo continuo.</li>
</ul>

<h4>Comandos especiales (IPython/mágicos)</h4>

<ul>
  <li><code>%run</code>: ejecuta un script Python
    <pre><code>%run ./miprograma.py</code></pre>
  </li>
  
  <li><code>%time</code>: mide tiempo de ejecución de una línea
    <pre><code>%time sum(range(1000000))</code></pre>
  </li>
  
  <li><code>%%time</code>: mide tiempo de ejecución de una celda completa
    <pre><code>%%time

  </li>
  
  <li><code>!</code>: ejecuta comando de consola/shell

!pip list</code></pre>

  <li><code>%%bash</code>: ejecuta múltiples comandos bash

ls -la
pwd
echo "Directorio actual mostrado"</code></pre>

  <li><code>%%js</code>: ejecuta código JavaScript

<h2>2.12 Performance Sample, NumPy vs Plain Python</h2>

```python
import numpy as np
my_arr = np.arange(10000000) # Utilizando NumPy Arrays
my_list = list(range(10000000)) # Utilizando Listas

print("Primeros 10 elementos:", my_list[:10])
print("Últimos 10 elementos:", my_list[-10:])
print("Longitud de la lista:", len(my_list))
```

```python
%time for _ in range(10): my_arr2 = my_arr * 2
%time for _ in range(10): my_list2 = [x * 2 for x in my_list]
```
